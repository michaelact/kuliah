```python
# Import required libraries
import pandas as pd
```

```python
!wget https://raw.githubusercontent.com/rebekz/datascience_course/main/data/housing/housing.csv
df = pd.read_csv('housing.csv')
display(df)
```

```python
# Check NaN values in `total_bedrooms` column
df['total_bedrooms'].isnull().sum()
```

```python
# Get median values in `total_bedrooms`
median_total_bedrooms = df['total_bedrooms'].median()
median_total_bedrooms
```

```python
# Data Transformation (Filling NaN values with median)
df['total_bedrooms'].fillna(median_total_bedrooms)
```
